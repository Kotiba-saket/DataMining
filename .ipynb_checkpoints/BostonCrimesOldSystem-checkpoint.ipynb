```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *

crimes_schema = StructType([StructField("COMPNOS", StringType(), True),
                            StructField("NatureCode", StringType(), True),
                            StructField("INCIDENT_TYPE_DESCRIPTION", StringType(), True),
                            StructField("MAIN_CRIMECODE", StringType(), True),
                            StructField("REPTDISTRICT", StringType(), True),
                            StructField("REPORTINGAREA", StringType(), True),
                            StructField("FROMDATE", StringType(), True),
                            StructField("WEAPONTYPE", StringType(), True),
                            StructField("Shooting", StringType(), True),
                            StructField("DOMESTIC", StringType(), True),
                            StructField("SHIFT", StringType(), True),
                            StructField("Year", StringType(), True),
                            StructField("Month", IntegerType(), True),
                            StructField("DAY_WEEK", StringType(), True),
                            StructField("UCRPART", StringType(), True),
                            StructField("X", StringType(), True),
                            StructField("Y", StringType(), True),
                            StructField("STREETNAME", StringType(), True),
                            StructField("XSTREETNAME", StringType(), True),
                            StructField("Location", StringType(), True)
])
                            
spark = SparkSession.builder.appName("Crimes in boston Old").getOrCreate()
crimes = spark.read.csv("./data/BostonCrimesOld.csv", header = True, schema = crimes_schema)
```

```python
crimes
```

```python
crimes.groupBy("YEAR").count().orderBy("YEAR").show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_axis = [i+1 for i in range(11)]
y_axis = []

result = crimes.groupBy("MONTH").count().orderBy("MONTH")
data = result.take(12)
for i in range(1,12):
    y_axis.append(data[i-1][1])

fig1 = plt.figure(figsize=(7, 7))
fig1.suptitle('Crimes per month per within a hour', fontsize=14)
ax = fig1.add_subplot(111)
ax.set_xlabel('month')
ax.set_ylabel('Crime')
ax.bar(x_axis, y_axis, fc='darksalmon', align='center')
plt.show()
```

```python
result = crimes.groupBy("REPTDISTRICT").count().orderBy("REPTDISTRICT")
result.show()
```

```python
location = [str(item[0]) for item in result.limit(20).collect()]
count = [item[1] for item in result.limit(20).collect()]

x_axis = [i for i in location] 
y_axis= [i for i in count]

fig = plt.figure(figsize=(20,20))
fig.suptitle('All crimes in boston from 2012-15', fontsize=36)
 
plt.xlabel("Number of Crimes", fontsize = 28)
plt.ylabel("Crimes Location", fontsize = 28)
plt.barh(x_axis,y_axis,color = "red")
plt.xticks(size = 24)
plt.yticks(size = 24)
plt.show()
```

```python
crime_long_lat  = crimes.select("Location").limit(9999).collect()
str(crime_long_lat[0][0])
```

```python
def geoStrToLocation(geo):
    location = [0,0]
    splited = geo.split(',')
    location[0] = float(splited[0].replace('(',''))
    location[1] = float(splited[1].replace(')', ''))
    return location

location = geoStrToLocation('(42.34638135, -71.10379454)')
location
```

```python
test = [geoStrToLocation(item[0]) for item in crime_long_lat]
```

```python
from folium.plugins import HeatMap
import folium

yearMap = folium.Map(
    location=[42.34862382, -71.08277637],
    tiles='Stamen Toner',
    zoom_start=13
)
HeatMap(test,name=None, min_opacity=0.5, max_zoom=18, max_val=1.0, radius=25, blur=10, gradient=None, overlay=True, control=True, show=True).add_to(yearMap)
yearMap
```
